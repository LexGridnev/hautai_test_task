```python
import cv2
from matplotlib import pyplot as plt
%matplotlib inline

from segmentation import PixelClustering
```

```python
IMAGE_PATH = 'image/1.png'
```

## RGB Image without coordinates as a features

```python
image = cv2.cvtColor(cv2.imread(IMAGE_PATH), cv2.COLOR_BGR2RGB)
pixel_clustering = PixelClustering()
mask = pixel_clustering.fit_predict(image)
plt.imshow(mask)
None
```

## RGB image with pixel coordinates as a features in cartesian system

```python
image = cv2.cvtColor(cv2.imread(IMAGE_PATH), cv2.COLOR_BGR2RGB)
pixel_clustering = PixelClustering(coordinate_system='euclidean')
mask = pixel_clustering.fit_predict(image)
plt.imshow(mask)
None
```

## RGB image with pixel coordinates as a features in polar system

```python
image = cv2.cvtColor(cv2.imread(IMAGE_PATH), cv2.COLOR_BGR2RGB)
pixel_clustering = PixelClustering(coordinate_system='polar')
mask = pixel_clustering.fit_predict(image)
plt.imshow(mask)
None
```

## HSV image without coordinates as a features

```python
image = cv2.cvtColor(cv2.imread(IMAGE_PATH), cv2.COLOR_BGR2HSV)
pixel_clustering = PixelClustering()
mask = pixel_clustering.fit_predict(image)
plt.imshow(mask)
None
```

## HSV image with pixel coordinates as a features in cartesian system

```python
image = cv2.cvtColor(cv2.imread(IMAGE_PATH), cv2.COLOR_BGR2HSV)
pixel_clustering = PixelClustering(coordinate_system='euclidean')
mask = pixel_clustering.fit_predict(image)
plt.imshow(mask)
None
```

## HSV image with pixel coordinates as a features in polar system

```python
image = cv2.cvtColor(cv2.imread(IMAGE_PATH), cv2.COLOR_BGR2HSV)
pixel_clustering = PixelClustering(coordinate_system='polar')
mask = pixel_clustering.fit_predict(image)
plt.imshow(mask)
None
```

## H-component of HSV image with pixel coordinates as a features in polar system

```python
image = cv2.cvtColor(cv2.imread(IMAGE_PATH), cv2.COLOR_BGR2HSV)
pixel_clustering = PixelClustering(channel_indices=(0, ), coordinate_system='polar')
mask = pixel_clustering.fit_predict(image)
plt.imshow(mask)
None
```

## H-component of HSV image with pixel coordinates as a features in cartesian system

```python
image = cv2.cvtColor(cv2.imread(IMAGE_PATH), cv2.COLOR_BGR2HSV)
pixel_clustering = PixelClustering(channel_indices=(0, ), coordinate_system='euclidean')
mask = pixel_clustering.fit_predict(image)
plt.imshow(mask)
None
```

It seemed to me that there is not much difference between clustering in ordinary rgb space and HSV space, even when using pixel coordinates as additional features.
